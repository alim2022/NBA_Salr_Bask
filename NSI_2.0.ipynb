```python
import pandas as pd
import numpy as np
import math as mt
import matplotlib.pyplot as plt
```

```python
salariespoints_df = pd.read_csv("/Users/lima22/Desktop/NBA_Salr_Bask/nbasalariespoints.csv")
```

```python
Newdf = salariespoints_df[["Player","FG%","AST","BLK","STL","FTA","ORB","DRB","PF","TOV","3P","2P","FT","SALARY"]]
```

```python
Newdf = Newdf.dropna()
```

```python
Newdf["PTS"] = (Newdf["3P"] + Newdf["2P"] + 0.5*Newdf["FT"])
```

```python
Newdf["NSI_2.0"] = np.sqrt(10*Newdf["FG%"]) + Newdf["PTS"] + Newdf["AST"] + 0.8*Newdf["BLK"] + 1.2*Newdf["STL"] - (Newdf["FTA"] - Newdf["FT"]) + 0.6*Newdf["ORB"] + 0.4*Newdf["DRB"] - 0.4*Newdf["PF"] - 1.1*Newdf["TOV"]
```
