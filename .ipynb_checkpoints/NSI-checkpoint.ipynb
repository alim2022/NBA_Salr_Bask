```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
salariespoints_df = pd.read_csv("/Users/lima22/Desktop/NBA_Salr_Bask/nbasalariespoints.csv")
```

```python
Newdf["SALARY"] = Newdf["SALARY"].str.replace('$', '')
```

```python
Newdf["SALARY"] = Newdf["SALARY"].str.replace(',', '')
```

```python
Newdf = salariespoints_df.dropna()
```

```python
SIM = (pd.to_numeric(Newdf["SALARY"]))*0.000001
```

```python
Newdf["NSI"] = 10*(((100*(Newdf['2P%'] + Newdf['3P%'])) - (Newdf['TOV']+(0.5*(Newdf['PF'])))) + (0.65*(Newdf['AST'])) + (0.8* (Newdf['STL'] + Newdf['BLK'])))
```

```python
fig, ax = plt.subplots() 

ax.scatter(SIM, Newdf["NSI"], 
            alpha= 0.70) 

ax.set_title('Correlation between Salaries and Performance', loc='left', fontsize=14, fontweight = "bold") 

ax.set_xlabel("Salaries (by millions)")
ax.set_ylabel("NSI")

ax.spines["right"].set_visible(False) 
ax.spines["top"].set_visible(False) 

plt.show()
```
